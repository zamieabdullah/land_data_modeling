```python
import requests
import pandas as pd
```

```python
url = "https://econdata.s3-us-west-2.amazonaws.com/Reports/Core/RDC_Inventory_Core_Metrics_County_History.csv"

df = pd.read_csv(url)

# Step 2: Split the 'county_name' column into 'county' and 'state'
df[['county', 'state']] = df['county_name'].str.split(',', expand=True)

# Optionally: Strip any leading/trailing spaces from the new columns
df['county'] = df['county'].str.strip()
df['state'] = df['state'].str.strip()

# Step 3: Optionally drop the original 'county_name' column
df = df.drop(columns=['county_name'])

df[df['state'] == 'ny']
```
